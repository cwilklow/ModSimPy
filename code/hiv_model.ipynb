```python
def run_simulation(system, update_func):
    """Runs a simulation of the system.
        
    system: System object
    update_func: function that updates state
    
    returns: TimeFrame
    """
    
    frame = TimeFrame(columns=state.index)
    frame.row[t0] = state
    
    for t in linrange(t0, t_end):
        frame.row[t+1] = update_func(frame.row[t], t, system)
    
    return frame
```

```python
state=State(R=0, L=0,E=0,V=0)
'''R=uninfected CD4 lymphocytes
L=latently infected
E=actively infected
V=free virons'''
print(state)
```

```python
system=System(tau=0, gamma=0, mu=0, beta=0, alpha=0, delta=0,rho=0,pi=0,sigma=0)
'''gamma is rate that new uninfected CD4 lymphocytes arise, rate that R increases
tau is proportion activated
mu is HIV independent death rate of CD4 lymphocytes
rho is proportion that become latently infected
of this proportion alpha is the proportion activated
delta is death rate of actively infected cells
pi is rate of free virons produced
sigma is removal of infected cells'''
print(system)
```

```python
t0=0
t_end=24
t=linrange(t0,t_end,0.01)
```

```python
def update_func(state,t, system):
    '''Update the model to reflect addition and deletion of activated uninfected lymphocytes'''
    dR_dt=(system.gamma*system.tau-system.mu*state.R-system.beta*state.R*state.V)
    dL_dt=(system.rho*system.beta*state.R*state.V-system.mu*state.L-system.alpha*state.L)
    dE_dt=((1-system.rho)*system.beta*state.R*state.V+system.alpha*state.L-system.delta*state.E)
    dV_dt=(system.pi*state.E-system.sigma*state.V)
    
    state.R=state.R*dR_dt
    state.L=state.L*dL_dt
    state.E=state.E*dE_dt
    state.V=state.V*dV_dt
    
    return state
```

```python
run_simulation(system,update_func)
```
